```python
import logging
import extract
import pandas as pd
pd.set_option("display.max_rows", None, "display.max_columns", None)

logging.basicConfig(format='%(asctime)s %(message)s')
logger = logging.getLogger()
logger.setLevel(logging.INFO)

params = {"codeCommuneEtablissement": "*","etatAdministratifUniteLegale": "A",
'categorieJuridiqueUniteLegale': "5*",
'activitePrincipaleUniteLegale': "16.10A", 'etablissementSiege':True,}
file=extract.query(logger, params, True)
import myGoogleSearch

ts=myGoogleSearch.go(file,logger)
```

```python

ts[['denomination', 'siret','groupEffectif','effectifBrut', 'effectif']]
```

# Nombre d'entreprises par tranche de région

```python
import pandas as pd
pd.set_option("display.max_rows", None, "display.max_columns", None)

byRegion = pd.pivot_table(ts,index='region',
                                    aggfunc='count',
                                    values=['siret'],
                                    margins=True).sort_values(by=['siret'], ascending=False)
byRegion
```

# Par CA et effectif

```python
import plotly.express as px

df = ts[['CA_average','effectif', 'denomination', "departement"]]

fig = px.scatter(df, text=df["denomination"], x=df["CA_average"], y=df["effectif"], 
                 title="Par CA et Effectif", 
                 hover_name="denomination", hover_data=["CA_average", "effectif"],
                color=df["departement"])
fig.update_traces(mode='markers', marker=dict(
        size=16,
        
        colorscale='Viridis', showscale=False),
                  hovertemplate = "%{text} <br>CA : %{x:6,.0f} </br> Effectif : %{y}")
fig.update_layout(
    xaxis={
        'title': "Chiffre d'affaires (k€)"},
    yaxis={'title':'Effectif'})
fig.update_layout(showlegend=False)

fig.show()
```
