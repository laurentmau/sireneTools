```python
ld=[]
for d in ts['domains']:
    for i in d:
        ld.append(i)
frequency = {}
for item in ld:
   frequency[item] = ld.count(item)
freqSorted=dict(sorted(frequency.items(), key=lambda item: item[1], reverse=True))
for key, value in freqSorted.items():
   print("% s -> % d" % (key, value))

```

```python
ld=[]
dr=[]
for d in ts['ownedDomain']:
    ld.append(d)
frequency = {}
for item in ld:
   frequency[item] = ld.count(item)
freqSorted=dict(sorted(frequency.items(), key=lambda item: item[1], reverse=True))
for key, value in freqSorted.items():
    print("% s -> % d" % (key, value))
    if value >2:
        dr.append(key)
        
print (dr)    
```
